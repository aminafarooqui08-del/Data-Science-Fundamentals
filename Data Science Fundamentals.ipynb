1. Descriptive Statistics

Definition:

Descriptive Statistics summarizes and describes the main features of a dataset without drawing conclusions.

Mean = Average

Median = Middle value

Mode = Most frequent value

Variance = Spread of data

Standard Deviation = Square root of variance

Range = Max − Min

```python
import numpy as np

data = [10, 20, 30, 40, 50]

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
print("Minimum:", np.min(data))
print("Maximum:", np.max(data))
```

4. Sampling & Central Limit Theorem (CLT)

Definition:

Sampling means selecting a subset from a population.

```python
import numpy as np

population = np.random.randint(1,100,10000)

sample = np.random.choice(population,100)

print(np.mean(sample))
```

5. Hypothesis Testing

Definition:

Used to decide whether a claim about data is true.

```python
from scipy.stats import ttest_1samp

data = [12,15,14,13,12,15]

t,p = ttest_1samp(data,13)

print("P-value:",p)
```

6. Correlation & Covariance

Definition:

Covariance measures how two variables change together.  

Correlation measures both direction and strength of the relationship.

```python
import numpy as np

x = [1,2,3,4,5]
y = [2,4,6,8,10]

print(np.cov(x,y))
print(np.corrcoef(x,y))
```

9. Data Preprocessing (Scikit-Learn)

Definition:

Preparing raw data before feeding it into ML algorithms.

10. Regression

Definition:

Regression predicts continuous values.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])

model = LinearRegression()

model.fit(X,y)

print(model.predict([[5]]))
```

11. Classification

Definition:

Classification predicts categories.

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([[1],[2],[3],[4]])
y = np.array([0,0,1,1])

model = LogisticRegression()

model.fit(X,y)

print(model.predict([[2.5]]))
```

12. Clustering

Definition:

Groups similar data points without labels.

```python
from sklearn.cluster import KMeans
import numpy as np

X = np.array([[1,2],[1,3],[10,10],[11,11]])

kmeans = KMeans(n_clusters=2, random_state=42)

kmeans.fit(X)

print(kmeans.labels_)
```

13. PCA (Principal Component Analysis)

Definition:

PCA reduces the number of features while preserving as much information (variance) as possible.

```python
from sklearn.decomposition import PCA
import numpy as np

X = np.random.rand(10,5)

pca = PCA(n_components=2)

X_new = pca.fit_transform(X)

print(X_new.shape)
```

14. Model Evaluation

Definition:

Model Evaluation measures how well a machine learning model performs on unseen data.

```python
from sklearn.metrics import accuracy_score

y_true = [1,0,1,1]
y_pred = [1,0,0,1]

print("Accuracy:", accuracy_score(y_true,y_pred))
```

15. Hyperparameter Tuning

Definition:

Hyperparameter tuning is the process of finding the best model settings (hyperparameters) before training to improve performance.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

model = SVC()

params = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}

grid = GridSearchCV(model, params, cv=3)

# grid.fit(X_train, y_train)

print(grid)
```
